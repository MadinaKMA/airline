```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
df['Gender']=LE.fit_transform(df['Gender'])
df['Customer Type']=LE.fit_transform(df['Customer Type'])
df['Type of Travel']=LE.fit_transform(df['Type of Travel'])
df['Class']=LE.fit_transform(df['Class'])
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.figure(figsize=(11,11))
sns.histplot(data=df, x='Age', hue='Gender')
sns.set_theme(style='ticks')
plt.show()
```

```python
sns.set_theme(style='darkgrid')
```

```python
df.drop(['id'], axis=1, inplace=True)
df.corrwith(df['satisfaction']).abs().sort_values(ascending=False)*100
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
strat_split=StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for tr_index, ts_index in strat_split.split(df, df['Gender']):
  spt_trainset=df.loc[tr_index]
  spt_testset=df.loc[ts_index]
```

```python
X_train = spt_trainset.drop('satisfaction', axis=1)
y_train = spt_trainset['satisfaction'].copy()

X_test = spt_testset.drop('satisfaction', axis=1)
y_test = spt_testset['satisfaction'].copy()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OrdinalEncoder,StandardScaler,OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
pipeline = Pipeline([
            ('median_imputer', SimpleImputer(strategy='median')),
            ('std_scaler', StandardScaler())
])
x_train_prepared=pipeline.fit_transform(X_train)
x_test_prepared=pipeline.fit_transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score
f1=[]
for k in range(1,25):
  knn = KNeighborsClassifier(n_neighbors=k)
  knn.fit(x_train_prepared,y_train)
  y_predicted=knn.predict(x_test_prepared)
  f1.append(f1_score(y_test, y_predicted))
plt.figure(figsize=(11,11))
plt.plot(range(1,25), f1)
plt.xticks(range(1,25))
plt.show()
```

```python
from sklearn.metrics import (classification_report, accuracy_score,
                             confusion_matrix, roc_curve, auc, RocCurveDisplay)
knn_model = KNeighborsClassifier(n_neighbors=9)
knn_model.fit(x_train_prepared, y_train)

knn_y_predicted = knn_model.predict(x_test_prepared)

print(classification_report(y_test, knn_y_predicted))
print("Model aniqligi:", accuracy_score(y_test, knn_y_predicted))

conf=confusion_matrix(y_test, knn_y_predicted)
sns.heatmap(conf, annot=True, fmt='g')

fpr, tpr, thresholds = roc_curve(y_test, knn_y_predicted)
roc_auc = auc(fpr, tpr)
disp = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='Roc curve')
disp.plot();
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RF_model=RandomForestClassifier()
RF_model.fit(x_train_prepared, y_train)
predicted_test_data = RF_model.predict(x_test_prepared)

print('Model aniqligi: ', accuracy_score(y_test, predicted_test_data))
conf=confusion_matrix(y_test, predicted_test_data)
sns.heatmap(conf, annot=True, fmt='g')
```
